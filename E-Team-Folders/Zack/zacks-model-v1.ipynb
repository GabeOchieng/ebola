```python
data_file_str = '../../SharedData/data/ebola_guinea_liberia_SL_nigeria_cases_timeseries.csv'
df = pd.read_csv(data_file_str)
df['Date'] = pd.to_datetime(df['Date'])
df = df.set_index('Date')


diff_dict = defaultdict(list)
colz = df.columns

df2 = pd.DataFrame()
def cumdiff(df, col):
    return df[col] - df[col].shift(1)
for c in colz:
    df2[c] = cumdiff(df, c)
df2.plot(figsize=(13,8))
print df2.icol(1).squeeze().values
print df2.icol(1).squeeze().values
fig = plt.figure(figsize=(21,13))
ax1 = fig.add_subplot(211)

```

```python
data_file_str = '../../SharedData/data/ebola-data-db-format.csv'
df = pd.read_csv(data_file_str)

filter_text = "Cumulative number of confirmed Ebola cases"
condition = df.Indicator == filter_text

df = df.reset_index(drop=True)
df.index = pd.to_datetime(df['Date'])
daily_totals = df.groupby('Date')['value'].max()
daily_totals = daily_totals.sort_index()

daily_diffs = []
for i, cell in enumerate(daily_totals.values):
    daily_diffs.append(daily_totals[i+1] - cell)
    if i+2 >= len(daily_totals):
        break
# daily_diffs.insert(0, daily_totals[0])
daily_diffs.insert(0, 0)

df2 = pd.DataFrame( {'daily_totals': daily_totals, 'daily_diffs': daily_diffs} )
df2.daily_diffs.plot(figsize=(13,8))

fig = plt.figure(figsize=(12,8))
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(df2.daily_diffs.squeeze(), ax=ax1)
```
